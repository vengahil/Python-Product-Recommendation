```python
cities = [
    ('Coimbatore',11.0168,76.9558),
    ('Salem',11.6,78.2),
    ('Chennai',13.01,80.4)]

icons=[]
temps=[]
for city in cities:
    lat = city[1]
    lon = city[2]
    #line='https://7e1aa32e-7542-4f01-856b-b3ec6a73f0d3:nbxEB9NhPm@twcservice.eu-gb.mybluemix.net/api/weather/v1/geocode/'+str(lat)+'/'+str(lon)+'/observations.json?&units=m'
    line='https://d822b176-43af-45fd-93c7-dcc5d923c9dc:A8TNbKDgKN@twcservice.mybluemix.net/api/weather/v1/geocode/'+str(lat)+'/'+str(lon)+'/observations.json?&units=m'
    r=requests.get(line)
    weather = json.loads(r.text)    
    icons=np.append(icons,weather['observation']['wx_icon'])    
    temps=np.append(temps,weather['observation']['temp'])
```

```python
from mpl_toolkits.basemap import Basemap
from matplotlib.offsetbox import AnnotationBbox, OffsetImage
from matplotlib._png import read_png
#from itertools import izip # bug fixed as below
izip = zip
import urllib # .open bug fiaxed as below, .request.open
import urllib.request
import os
import sys
%config IPCompleter.greedy=True
```

```python
#tn map
#matplotlib.style.use('bmh')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 12))
 
# background maps
# c -coarse, i - intermedate, h- high , f - full None
m1 = Basemap(projection='merc',resolution='h',llcrnrlon=75.0,llcrnrlat=7.5,urcrnrlon=81.5,urcrnrlat=14.0,ax=axes[0])
m1.drawlsmask(land_color='dimgrey',ocean_color='dodgerBlue',lakes=True,resolution='h')
m1.drawcoastlines()
#m1.drawmapboundary(fill_color='aqua')
m1.drawrivers()
 
m2 = Basemap(projection='merc',resolution=None,llcrnrlon=75.0,llcrnrlat=7.5,urcrnrlon=81.5,urcrnrlat=14.0,ax=axes[1])
m2.drawlsmask(land_color='dimgrey',ocean_color='dodgerBlue',lakes=True)
 
# weather icons map
for [icon,city] in izip(icons,cities):
    lat = city[1]
    lon = city[2]
    #pngfile=urllib.request.urlopen('http://github.com/ibm-cds-labs/python-notebooks/blob/master/weathericons/icon'+str(int(icon))+'.png?raw=true')
    #fileandpath="icon"+str(int(icon))+".png"
    pngfile="icon"+str(int(icon))+".png"
    icon_hand = read_png(pngfile)
    imagebox = OffsetImage(icon_hand, zoom=.15)
    ab = AnnotationBbox(imagebox,m1(lon,lat),frameon=False) 
    axes[0].add_artist(ab)
 
# temperature colours   
jetcols = plt.get_cmap('jet')
 
# temperature map
for [temp,city] in izip(temps,cities):
    lat = city[1]
    lon = city[2]
    # normalize temperature between 0 and 40 degrees
    tempnorm = (temp-0.0)/(40.0-0.0)
    x1, y1 = m2(lon,lat)
    bbox_props = dict(boxstyle="round,pad=0.3", fc=jetcols(tempnorm), ec=jetcols(tempnorm), lw=2)
    axes[1].text(x1, y1, temp, ha="center", va="center",
                size=11,bbox=bbox_props)
     
plt.tight_layout()
```
